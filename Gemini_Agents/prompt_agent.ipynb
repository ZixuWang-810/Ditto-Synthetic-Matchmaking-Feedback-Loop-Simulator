```python
import os
from gemini_agent import Agent
from IPython.display import display, Markdown
```

```python
role = 'You are a professional prompt engineer specializing in designing clear, precise, and production-ready prompts.'

system_prompt = """ As a professional prompt engineer
    Your task is to construct high-quality role-based prompts tailored to the user’s needs.
    
    Everytime when you build a new agent, you need to follow these rules:
        1. [Role]
            - Define the exact persona or expertise framing.
        2. [Task]
            - Provide explicit, unambiguous instructions.
        3. [Output Format]
            - Clearly specify structure, style, and constraints.
        4. [Examples] (when beneficial)
            - Provide representative input/output demonstrations.
        5. [Context]
            - Include assumptions, constraints, or environmental details.

    Quality Requirements:
        - Maximize clarity and specificity
        - Minimize ambiguity
        - Reduce hallucination risk
        - Prefer actionable instructions
        - Avoid unnecessary verbosity

    Interaction Rules:
        - If the user’s request is underspecified, ask concise clarifying questions
        - Otherwise, make reasonable assumptions and state them in [Context]
        - When useful, provide multiple prompt variants (e.g., concise vs robust)
    
"""
prompt_agent = Agent(role=role, system_prompt=system_prompt)
# Test the agent
prompt_agent.chat("What is your job?")
```

```python
# Create a system engineer prompt
request_system_engineer = """
Design a role-based prompt that creates a System Design Engineer Agent.
The agent should specialize in designing detailed project roadmaps for new projects, primarily focused on:
    - Software development
    - AI / Machine Learning systems
The agent must:
    - Produce structured, step-by-step development roadmaps
    - Consider real-world constraints (time, resources, technical risk, dependencies)
    - Optimize for efficiency and practicality
    - Follow industry best practices and engineering standards
The roadmap should include:
    - Problem definition & requirement analysis
    - Architecture & technology selection
    - Development phases & milestones
    - Risk analysis & trade-offs
    - Scalability & maintainability considerations
    - Deployment & iteration strategy
The prompt you design should:
    - Clearly define the agent’s role & expertise
    - Specify reasoning depth & decision-making style
    - Enforce structured outputs
    - Minimize vague or generic responses
    - Encourage practical engineering thinking over theoretical discussion
    - Output a production-quality prompt suitable for building this agent.
"""
promt = prompt_agent.chat(request_system_engineer)

```

```python
display(Markdown(promt))
```

```python
# Create a software developer
request_software_developer = """
Design a role-based prompt that creates a Software Development Engineer Agent.
The agent’s responsibility is to implement projects by strictly following the roadmap and architecture defined by a System Design Engineer.

The Software Development Engineer Agent must:
- Translate system design documents into production-quality code
- Follow architectural constraints and technology stack decisions
- Write professional, efficient, and maintainable code
- Adhere to industry best practices and coding standards
- Consider performance, scalability, and reliability
- Minimize technical debt

The agent should:
- Ask clarifying questions if design specifications are ambiguous
- Explicitly state assumptions when necessary
- Follow modular design principles
- Include meaningful comments and documentation
- Consider testing, validation, and error handling
- Optimize for readability and long-term maintainability

The prompt you design must:
- Clearly define the agent’s expertise level (e.g., senior software engineer)
- Enforce structured outputs (e.g., explanation → architecture alignment → implementation → tests)
- Specify coding quality standards
- Encourage constraint-aware implementation
- Prevent generic or low-quality code generation

The agent should prioritize:
- Code correctness
- Efficiency
- Robust error handling
- Clean structure
- Production readiness

Output a production-ready system prompt suitable for building this agent."""
sde_prompt = prompt_agent.chat(request_software_developer)
```

```python
display(Markdown(sde_prompt))
```
